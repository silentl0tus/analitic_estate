```python
### АНАЛИЗ ЦЕНЫ (ИСПРАВЛЕННЫЙ КОД) ###
print("=== АНАЛИЗ ЦЕНЫ ===")

# Смотрим, какие бывают записи в колонке "Цена"
print("Примеры значений в колонке 'Цена':")
print(df['Цена'].head(3).values)

# Пробуем извлечь число из цены (ИСПРАВЛЕННЫЙ СПОСОБ)
# Вариант 1: Простой способ (может не сработать для всех форматов)
try:
    df['price_number'] = df['Цена'].astype(str).str.extract(r'(\d+\.?\d*)')[0].astype(float)
    print("Способ 1 сработал!")
except:
    print("Способ 1 не сработал, пробуем способ 2...")
    # Вариант 2: Более надежный способ
    df['price_number'] = df['Цена'].apply(lambda x: float(str(x).split()[0]) if pd.notna(x) else np.nan)

print(f"\nУдалось извлечь цены: {df['price_number'].notna().sum()} из {len(df)}")

# Проверяем, создалась ли колонка
if 'price_number' in df.columns:
    # Простые статистики по цене
    print("\n=== СТАТИСТИКА ПО ЦЕНАМ ===")
    print(f"Минимальная цена: {df['price_number'].min():.0f} руб.")
    print(f"Максимальная цена: {df['price_number'].max():.0f} руб.")
    print(f"Средняя цена: {df['price_number'].mean():.0f} руб.")
    print(f"Медианная цена (типичная): {df['price_number'].median():.0f} руб.")
else:
    print("Колонка 'price_number' не создалась!")
    # Показываем первые несколько значений, чтобы понять формат
    print("\nПервые 5 значений колонки 'Цена':")
    for i, value in enumerate(df['Цена'].head().values):
        print(f"{i}: {value}")
```

```python
# Простые статистики по цене
print("\n=== СТАТИСТИКА ПО ЦЕНАМ ===")
print(f"Минимальная цена: {df['price_number'].min():.0f} руб.")
print(f"Максимальная цена: {df['price_number'].max():.0f} руб.")
print(f"Средняя цена: {df['price_number'].mean():.0f} руб.")
print(f"Медианная цена (типичная): {df['price_number'].median():.0f} руб.")
```

```python
if 'price_number' in df.columns and df['price_number'].notna().any():
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    # Гистограмма
    ax1.hist(df['price_number'].dropna(), bins=50, color='blue', alpha=0.7, edgecolor='black')
    ax1.set_xlabel('Цена (руб.)')
    ax1.set_ylabel('Количество объявлений')
    ax1.set_title('Распределение цен на аренду')
    ax1.grid(True, alpha=0.3)
    # Boxplot
    ax2.boxplot(df['price_number'].dropna())
    ax2.set_title('Выбросы в ценах')
    ax2.set_ylabel('Цена (руб.)')
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
```

```python
top_prices = df['price_number'].value_counts().head(10)
plt.figure(figsize=(10, 5))
top_prices.plot(kind='bar', color='green', alpha=0.7)
plt.title('Топ-10 самых частых цен аренды')
plt.xlabel('Цена (руб.)')
plt.ylabel('Количество объявлений')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

```python
# 3. График пропусков (всегда работает)
print("\n=== ГРАФИК: ПРОПУСКИ В ДАННЫХ ===")

# Считаем процент пропусков
missing_percent = (df.isna().sum() / len(df)) * 100
missing_percent = missing_percent[missing_percent > 0].sort_values(ascending=False)

plt.figure(figsize=(12, 6))
bars = plt.bar(range(len(missing_percent)), missing_percent.values, color='red', alpha=0.7)
plt.title('Процент пропущенных значений по колонкам', fontsize=14)
plt.ylabel('Процент пропусков (%)', fontsize=12)
plt.xlabel('Колонки', fontsize=12)
plt.xticks(range(len(missing_percent)), missing_percent.index, rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')

# Добавляем значения на столбцы
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{height:.1f}%', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# 4. Простая статистика по пропускам
print("\n=== СТАТИСТИКА ПО ПРОПУСКАМ ===")
print("Топ-5 колонок с наибольшими пропусками:")
for col, percent in missing_percent.head().items():
    print(f"{col}: {percent:.1f}% пропусков")
```
